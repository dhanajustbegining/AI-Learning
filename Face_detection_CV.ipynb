```python
# import the necessary libraries
import cv2 # image handling
import urllib.request # to import library from external URL
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
import io
from google.colab.patches import  cv2_imshow
```

```python
#download the pretrained clasifier (code to detect face) Haar cascade Classifier for face detection
url='https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'
urllib.request.urlretrieve(url,'haarcascade_frontalface_default.xml')
```

```python
# load face detection model
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
#load the image
#image=cv2.imread('/WhatsApp Image 2025-08-10 at 9.57.34 PM.jpeg')
image=cv2.imread('/content/sample_data/img.jpeg')
cv2_imshow(image)

```

```python
#black abs white -gray scale
grayscale_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale_img)
```

```python
grayscale_img.shape #estimation of length width and color channel
```

```python
# detect faces using face_cascade
faces=face_cascade.detectMultiScale(grayscale_img, scaleFactor=1.1, minNeighbors=1, minSize=(15,15))
```

```python
faces
```

```python
#draw faces on image
face_count=0
for (x,y,w,h) in faces :
  #drw rectangles on face locations
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2) #2 is thickness of rectangle

  #face_count+=1
  #extracting faces ROI region of interest
  face_crop=image[y:y+h,x:x+w]

  #sacw the extracted faces
  face_filename=f'face_{face_count}.jpg'
  cv2.imwrite(face_filename,face_crop)
  face_count=face_count+1
```
